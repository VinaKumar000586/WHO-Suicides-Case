# Visualization of Suicides data :
This dataset presents comprehensive statistics concerning self-inflicted injuries and suicides worldwide across a span of 118 countries. It encompasses various age groups, ranging from 0-85+ Years, for males, females, and individuals of unknown gender.

The data span from the year 1950 to 2021, although analysis suggests that the information for recent years might be incomplete. The data was sourced from the World Health Organization's (WHO) website. It's important to note that the rights pertaining to this data are subject to WHO's Copyright Policies, especially for any commercial uses


## Purpose of this Notebook:
In this notebook, my aim is to conduct an analysis focusing on several key aspects:

1. **Region Analysis**: I will delve into the data to gain insights on suicide rates across different regions.

2. **Country Analysis**: I intend to explore suicide trends within individual countries.

3. **Gender Analysis**: I'll examine suicide statistics based on gender.

4. **Year-Range Analysis**: Additionally, I will investigate the year-ranges that exhibit the highest rates of attempted suicides.

Through these analyses, I aim to uncover meaningful patterns, trends, and insights that shed light on the complex issue of suicide and self-inflicted injuries.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import plotly.express as px
```

```python
data=pd.read_csv('/kaggle/input/who-suicide-data-1950-2021/combined_processed_data.csv')
```

```python
data['Region Name'].unique()
```

In this dataset, there are six distinct regions: **'Africa', 'Asia', 'Central and South America', 'Europe', 'North America and the Caribbean', and 'Oceania'**

First of all, I will start the analysis with the data from the Asia region. In this section, I will attempt to identify the countries with the highest number of suicide cases, determine the age ranges with the highest suicide attempts, and compare the data between males and females.

```python
# Asia Region Data extract
Asia=data[data['Region Name']=='Asia']
```

```python
# groupby Asia data by Region Name, Country Name, and Sex
group_all_asia=Asia.groupby(by=['Region Name', 'Country Name','Sex']).sum()
# Reset index to convert multi-index to columns
group_all_asia.reset_index(inplace=True)
```

```python
# Create a stacked bar plot using Plotly Express
fig = px.bar(group_all_asia, x='Country Name', y='No of Suicides', color='Sex', title='Suicide Count by Country and Sex in Asia Region',
             labels={'No of Suicides': 'Number of Suicides'})

# Update layout for better visualization
fig.update_layout(barmode='stack', xaxis_tickangle=-45, xaxis=dict(categoryorder='total descending'))

# Show the plot
fig.show()
```

```python
# Asia Data separeted by Sex wise
Asia_male=group_all_asia[group_all_asia['Sex']=='Male']
Asia_female=group_all_asia[group_all_asia['Sex']=='Female']
```

## Asia Male

```python
Asia_male
```

```python
Asia_male.columns
```

```python
# Extract the columns related to year ranges
year_range_columns = Asia_male.columns[4:-3]
year_range_columns
```

```python
#New datafarme create for most suicidal year range and max finding
# Find the year range with the maximum value and its corresponding value for each country
Asia_male['Most Suicidal Year Range'] = Asia_male[year_range_columns].idxmax(axis=1)
Asia_male['Max Suicides in Year Range'] = Asia_male[year_range_columns].max(axis=1)

Male = Asia_male[['Country Name', 'Most Suicidal Year Range', 'Max Suicides in Year Range']]

print("For each country in region of Asia,the year range with the highest total suicides and its value for (male):")
Male
```

```python
# Create barplot for male
# Get the maximum and minimum values from 'Max Suicides in Year Range'
max_value = Male['Max Suicides in Year Range'].max()
min_value = Male['Max Suicides in Year Range'].min()

# Create a color map based on maximum and minimum values
color_map = plt.cm.get_cmap('viridis')

# Normalize the data for color mapping
norm = plt.Normalize(vmin=min_value, vmax=max_value)

# Create a bar plot
plt.figure(figsize=(12, 8))
bars = plt.barh(Male['Country Name'], Male['Max Suicides in Year Range'], color=color_map(norm(Male['Max Suicides in Year Range'])))

plt.xlabel('Number of Suicides')
plt.ylabel('Country Name')
plt.title('Most Suicidal Year Range and its Value for Each Country in Asia [Male]')
plt.tight_layout()

# Add data labels
for bar, value, year_range in zip(bars, Male['Max Suicides in Year Range'], Male['Most Suicidal Year Range']):
    plt.text(value, bar.get_y() + bar.get_height() / 2,f"{value} ({year_range})", va='center', color='green')


# Add color bar
sm = plt.cm.ScalarMappable(cmap=color_map, norm=norm)
sm.set_array([])  # Important for color bar to work
cbar = plt.colorbar(sm, orientation='vertical')
cbar.set_label('Max Suicides in Year Range')

plt.show()
```

```python
# pie plot
# Combine country name and year range for labels
labels = [f"{country} - {year_range}" for country, year_range in zip(Asia_male['Country Name'], Asia_male['Most Suicidal Year Range'])]

# Create a sunburst chart using Plotly Express
fig = px.sunburst(Asia_male, path=['Country Name', 'Most Suicidal Year Range'], values='Max Suicides in Year Range',
                  labels={'Max Suicides in Year Range': 'Count'}, hover_data=['Max Suicides in Year Range'])

# Update layout for better visualization
fig.update_layout(margin=dict(t=30, l=0, r=0, b=0), title='In Asia Region: Most Suicidal Year Range and Its Count by Country -> [Male]')

# Show the plot
fig.show()
```

```python
## function create for all data analysis

def most_suicidal_data_with_range(data,text):
    # Find the year range with the maximum value and its corresponding value for each country
    data['Most Suicidal Year Range'] = data[year_range_columns].idxmax(axis=1)
    data['Max Suicides in Year Range'] = data[year_range_columns].max(axis=1)

    result = data[['Country Name', 'Most Suicidal Year Range', 'Max Suicides in Year Range']]

    print(f"For each country in region of {text[0]}, the year range with the highest total suicides and its value for {text[1]}:")
    return result
    
    
    
    
```

```python
def barplot_for_region(data, text):
    # Get the maximum and minimum values from 'Max Suicides in Year Range'
    max_value = data['Max Suicides in Year Range'].max()
    min_value = data['Max Suicides in Year Range'].min()

    # Create a color map based on maximum and minimum values
    color_map = plt.cm.get_cmap('viridis')

    # Normalize the data for color mapping
    norm = plt.Normalize(vmin=min_value, vmax=max_value)

    # Create a bar plot
    plt.figure(figsize=(12, 8))
    bars = plt.barh(data['Country Name'], data['Max Suicides in Year Range'], color=color_map(norm(data['Max Suicides in Year Range'])))

    plt.xlabel('Number of Suicides')
    plt.ylabel('Country Name')
    plt.title(f'Most Suicidal Year Range and its Value for Each Country in {text[0]} [{text[1]}]')
    plt.tight_layout()

    # Add data labels
    for bar, value, year_range in zip(bars, data['Max Suicides in Year Range'], data['Most Suicidal Year Range']):
        plt.text(value, bar.get_y() + bar.get_height() / 2,f"{value} ({year_range})", va='center', color='green')


    # Add color bar
    sm = plt.cm.ScalarMappable(cmap=color_map, norm=norm)
    sm.set_array([])  # Important for color bar to work
    cbar = plt.colorbar(sm, orientation='vertical')
    cbar.set_label('Max Suicides in Year Range')

    plt.show()
```

```python
def pieplot_for_region(data,text):
    # Combine country name and year range for labels
    labels = [f"{country} - {year_range}" for country, year_range in zip(data['Country Name'], data['Most Suicidal Year Range'])]

    # Create a sunburst chart using Plotly Express
    fig = px.sunburst(data, path=['Country Name', 'Most Suicidal Year Range'], values='Max Suicides in Year Range',
                      labels={'Max Suicides in Year Range': 'Count'}, hover_data=['Max Suicides in Year Range'])

    # Update layout for better visualization
    fig.update_layout(margin=dict(t=30, l=0, r=0, b=0), title=f'In {text[0]} Region: Most Suicidal Year Range and Its Count by Country -> [{text[1]}]')

    # Show the plot
    fig.show()
```
